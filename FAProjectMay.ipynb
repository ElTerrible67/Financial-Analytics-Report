# EURJPY VS BITCOIN

```r
ohlc_eurjpy <- EUR_JPY[,c("OPEN","HIGH","LOW","CLOSE")]
```

```r
EURJPY_volatility_Garman <- volatility(ohlc_eurjpy, calc="garman")

Garman_frame_eurjpy <- as.data.frame(EURJPY_volatility_Garman)

Garman_frame_eurjpy <- Garman_frame_eurjpy[rowSums(is.na(Garman_frame_eurjpy)) == 0,]

rmse(Garman_frame,Garman_frame_eurjpy)
```

## Comparing the Yang Volatility for Bitcoin and EurJpy

```r
EURJPY_volatility_yang <- volatility(ohlc_eurjpy, calc="yang.zhang")
bitcoin_volatility_yang <- volatility(ohlc, calc="yang.zhang")


EURJPY_volatility_yang_df <- as.data.frame(EURJPY_volatility_yang)
EURJPY_volatility_yang_df <- EURJPY_volatility_yang_df[rowSums(is.na(EURJPY_volatility_yang_df)) == 0,]

bitcoin_volatility_yang_df <- as.data.frame(bitcoin_volatility_yang)
bitcoin_volatility_yang_df  <- bitcoin_volatility_yang_df [rowSums(is.na(bitcoin_volatility_yang_df )) == 0,]

rmse(EURJPY_volatility_yang_df,bitcoin_volatility_yang_df)
```
